```python
# Import necessary packages
import numpy as np
import pandas as pd
import os
import pickle
```

```python
y = pickle.load(open("team_list_y_namenum_non_ave_odds.p", "rb"))
player_dict = pickle.load(open("player_to_team_dict.p", "rb"))
player_map_dict = pickle.load(open("player_team_dict.p", "rb"))
team_dict = pickle.load(open("team.p", "rb"))
```

```python
# start_date = '10/19/2021'
# end_date = '04/15/2022'
start_date = '10/18/2022'
end_date = '04/09/2023'
date_list = [d.strftime('%m-%d-%Y') for d in pd.date_range(start=start_date, end=end_date)]
```

```python
player_map = list(player_map_dict.keys())
player_ind_dict = dict(zip(range(len(player_map)), player_map))
```

```python
team_map = list(team_dict.values())
team_ind_dict = dict(zip(range(len(team_map)), team_map))
```

```python
player_in_team = dict.fromkeys(list(player_dict.keys()))
for key in player_in_team:
  player_in_team[key] = ([], [])

assert len(date_list) == len(y)
for i in range(len(y)):
  key_list = y[i].keys()
  for key in key_list:
    for player in y[i][key][1]:
      if key[0] not in player_in_team[player][1]:
        player_in_team[player][0].append(date_list[i])
        player_in_team[player][1].append(key[0])
    for player in y[i][key][2]:
      if key[1] not in player_in_team[player][1]:
        player_in_team[player][0].append(date_list[i])
        player_in_team[player][1].append(key[1])
```

```python
c = 0
for player in player_in_team:
  if len(player_in_team[player][1]) > 1:
    c += 1
    s = []
    s.append(player_ind_dict[player])
    s.append("\n")
    flag = False
    for team_id in range(len(player_in_team[player][1])):
      if flag:
        s.append("->")
        s.append(player_in_team[player][0][team_id])
        s.append("->")
      s.append(team_ind_dict[player_in_team[player][1][team_id]])
      flag = True
    print("".join(s))
print("===", c, "===")
```

```python
team_record = dict.fromkeys(range(len(y)))
for key in team_record:
  team_record[key] = []

assert len(date_list) == len(y)
for i in range(len(y)):
  team_today = dict.fromkeys(list(team_ind_dict.keys()))
  key_list = y[i].keys()
  for key in key_list:
    team_today[key[0]] = y[i][key][1]
    team_today[key[1]] = y[i][key][2]
  team_record[i] = team_today
```

```python
order_list = list(team_record.keys())
order_list.reverse()
team_mem = dict.fromkeys(list(team_ind_dict.keys()))
for i in order_list:
  for t in team_record[i].keys():
    if team_record[i][t] is not None:
      team_mem[t] = team_record[i][t]
    else:
      team_record[i][t] = team_mem[t]
```

```python
pickle.dump(team_record, open("member_schedule.p", "wb"))
```
