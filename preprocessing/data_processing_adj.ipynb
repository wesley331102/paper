## Pass and Ast

```python
# settings
name_list = list(team_dict.values())
all_dict = dict.fromkeys(name_list)
date_list = [d.strftime('%m-%d-%Y') for d in pd.date_range(start=start_date, end=break_date)]

# iter
for date_string in date_list:
  pass_path = os.path.join('pass', date_string)
  files = os.listdir(pass_path)
  for file in files:
    team = file.split('.')[0]
    if all_dict[team] is None:
      all_dict[team] = {'pass': dict(), 'ast': dict(), 'match': 0, 'player': list()}
    all_dict[team]['match'] += 1
    df = pd.read_csv(os.path.join('pass', date_string, file))
    all_dict[team]['player'] = list(set(df['passer']) | set(df['receiver']) | set(all_dict[team]['player']))
    pass_key = all_dict[team]['pass'].keys()
    ast_key = all_dict[team]['ast'].keys()
    for _, row in df.iterrows():
      if (row['passer'], row['receiver']) not in pass_key:
        all_dict[team]['pass'][(row['passer'], row['receiver'])] = row['passes']
      else:
        all_dict[team]['pass'][(row['passer'], row['receiver'])] += row['passes']
      if (row['passer'], row['receiver']) not in ast_key:
        all_dict[team]['ast'][(row['passer'], row['receiver'])] = row['ast']
      else:
        all_dict[team]['ast'][(row['passer'], row['receiver'])] += row['ast']
```

```python
for team in name_list:
  game_match = float(all_dict[team]['match'])
  all_dict[team]['pass'].update((key, value / game_match) for key, value in all_dict[team]['pass'].items())
  all_dict[team]['ast'].update((key, value / game_match) for key, value in all_dict[team]['ast'].items())
```

```python
all_player_list = list()
for team in name_list:
  all_player_list = list(set(all_player_list) | set(all_dict[team]['player']))
all_player_list = [x for x in all_player_list if x == x]

pass_df = pd.DataFrame(np.zeros(shape=(len(all_player_list), len(all_player_list))), columns=all_player_list, index=all_player_list)
ast_df = pd.DataFrame(np.zeros(shape=(len(all_player_list), len(all_player_list))), columns=all_player_list, index=all_player_list)

for team in name_list:
  pass_data = all_dict[team]['pass']
  ast_data = all_dict[team]['ast']
  for key in pass_data:
    pass_df.loc[key[0], key[1]] = pass_data[key]
  for key in pass_data:
    ast_df.loc[key[0], key[1]] = ast_data[key]

  if not os.path.exists('output'):
    os.mkdir('output')

pass_df.to_csv(os.path.join('output', 'pass.csv'))
ast_df.to_csv(os.path.join('output', 'ast.csv'))
```

## Defend and Block

```python
# settings
def_df = pd.DataFrame(np.zeros(shape=(len(all_player_list), len(all_player_list))), columns=all_player_list, index=all_player_list)
stl_df = pd.DataFrame(np.zeros(shape=(len(all_player_list), len(all_player_list))), columns=all_player_list, index=all_player_list)
blk_df = pd.DataFrame(np.zeros(shape=(len(all_player_list), len(all_player_list))), columns=all_player_list, index=all_player_list)
pf_df = pd.DataFrame(np.zeros(shape=(len(all_player_list), len(all_player_list))), columns=all_player_list, index=all_player_list)
name_map = dict()
pos_map = dict()

# iter
# for date_string in date_list[:1]:
for date_string in date_list:
  match_list = schedule_dict[date_string.replace("-", "/")]
  # for match in match_list[:1]:
  for match in match_list:
    team_id1 = list(match.keys())[0]
    team_id2 = list(match.keys())[1]

    def_df_1 = pd.read_csv(os.path.join('def', date_string, str(team_id1) + '.csv'))
    data_df_1 = pd.read_csv(os.path.join('data', date_string, team_dict[team_id1]+'.csv'))
    df_1 = def_df_1.merge(data_df_1, left_on='CLOSE_DEF_PERSON_ID', right_on='PLAYER_ID')
    df_1 = df_1[['PLAYER_NAME_x', 'PLAYER_POSITION', 'D_FG_PCT', 'PCT_PLUSMINUS', 'PLAYER_ID', 'STL', 'BLK', 'BLKA', 'PF', 'PFD', 'PLUS_MINUS']]
    def_df_2 = pd.read_csv(os.path.join('def', date_string, str(team_id2) + '.csv'))
    data_df_2 = pd.read_csv(os.path.join('data', date_string, team_dict[team_id2]+'.csv'))
    df_2 = def_df_2.merge(data_df_2, left_on='CLOSE_DEF_PERSON_ID', right_on='PLAYER_ID')
    df_2 = df_2[['PLAYER_NAME_x', 'PLAYER_POSITION', 'D_FG_PCT', 'PCT_PLUSMINUS', 'PLAYER_ID', 'STL', 'BLK', 'BLKA', 'PF', 'PFD', 'PLUS_MINUS']]

    # name
    for index, row in df_1.iterrows():
      name = row['PLAYER_NAME_x'].split()
      if row['PLAYER_NAME_x'] not in name_map.keys():
        for player in all_player_list:
          # print(player)
          if all(x in player for x in name):
            name_map[row['PLAYER_NAME_x']] = player
            break

    for index, row in df_2.iterrows():
      name = row['PLAYER_NAME_x'].split()
      if row['PLAYER_NAME_x'] not in name_map.keys():
        for player in all_player_list:
          if all(x in player for x in name):
            name_map[row['PLAYER_NAME_x']] = player
            break

    # calculate adj by posotion(G, F, C)
    # first
    for index_1, player_1 in df_1.iterrows():
      position = [x for x in ['G', 'F', 'C'] if x in player_1['PLAYER_POSITION']]
      for index_2, player_2 in df_2.iterrows():
        if any(p in position for p in player_2['PLAYER_POSITION']):
          if (name_map[player_1['PLAYER_NAME_x']], name_map[player_2['PLAYER_NAME_x']]) not in pos_map.keys():
            pos_map[(name_map[player_1['PLAYER_NAME_x']], name_map[player_2['PLAYER_NAME_x']])] = 1
          else:
            pos_map[(name_map[player_1['PLAYER_NAME_x']], name_map[player_2['PLAYER_NAME_x']])] += 1

          def_score = float(max(player_1['PCT_PLUSMINUS'], 0))
          def_df.loc[name_map[player_1['PLAYER_NAME_x']], name_map[player_2['PLAYER_NAME_x']]] += def_score
          stl_score = float(player_1['STL'])
          stl_df.loc[name_map[player_1['PLAYER_NAME_x']], name_map[player_2['PLAYER_NAME_x']]] += stl_score
          blk_score = float(max(player_1['BLK']-player_1['BLKA'], 0))
          blk_df.loc[name_map[player_1['PLAYER_NAME_x']], name_map[player_2['PLAYER_NAME_x']]] += blk_score
          pf_score = float(max(player_1['PFD']-player_1['PF'], 0))
          pf_df.loc[name_map[player_1['PLAYER_NAME_x']], name_map[player_2['PLAYER_NAME_x']]] += pf_score

for key in pos_map.keys():
  def_df.loc[key[0], key[1]] = def_df.loc[key[0], key[1]] / pos_map[key]
  stl_df.loc[key[0], key[1]] = stl_df.loc[key[0], key[1]] / pos_map[key]
  blk_df.loc[key[0], key[1]] = blk_df.loc[key[0], key[1]] / pos_map[key]
  pf_df.loc[key[0], key[1]] = pf_df.loc[key[0], key[1]] / pos_map[key]

def_df.to_csv(os.path.join('output', 'def.csv'))
stl_df.to_csv(os.path.join('output', 'stl.csv'))
blk_df.to_csv(os.path.join('output', 'blk.csv'))
pf_df.to_csv(os.path.join('output', 'pf.csv'))
```

```python
# settings
name_list = list(team_dict.values())
history_dict = dict()
match_dict = dict()
team_df = pd.DataFrame(np.zeros(shape=(len(name_list), len(name_list))), columns=name_list, index=name_list)

# iter
for date_string in date_list:
  match_list = schedule_dict[date_string.replace("-", "/")]
  for match in match_list:
    teams = list(match.keys())
    team_id1 = teams[0]
    team_id2 = teams[1]
    if (team_dict[team_id1], team_dict[team_id2]) in history_dict.keys():
      history_dict[(team_dict[team_id1], team_dict[team_id2])] += 1
    elif (team_dict[team_id2], team_dict[team_id1]) in history_dict.keys():
      history_dict[(team_dict[team_id2], team_dict[team_id1])] += 1
    else:
      history_dict[(team_dict[team_id1], team_dict[team_id2])] = 1

    if match[team_id1] > 0:
      df_t1 = pd.read_csv(os.path.join('team', date_string, team_dict[team_id1]+'.csv'))
      team_df.loc[team_dict[team_id1], team_dict[team_id2]] += df_t1.loc[0, 'PLUS_MINUS']
    else:
      df_t2 = pd.read_csv(os.path.join('team', date_string, team_dict[team_id2]+'.csv'))
      # print(df_t2)
      team_df.loc[team_dict[team_id2], team_dict[team_id1]] += df_t2.loc[0, 'PLUS_MINUS']

for key in history_dict.keys():
  team_df.loc[key[0], key[1]] = team_df.loc[key[0], key[1]] / history_dict[key]
  team_df.loc[key[1], key[0]] = team_df.loc[key[1], key[0]] / history_dict[key]

team_df.to_csv(os.path.join('output', 'team.csv'))
```

# Normalize

```python
import scipy.sparse as sp
import numpy as np
import pandas as pd
import os
```

```python
def normalize_adj(adj, symmetric=True):
    if symmetric:
        d = sp.diags(np.power(np.array(adj.sum(1)), -0.5).flatten())
        a_norm = adj.dot(d).transpose().dot(d).tocsr()
    else:
        d = sp.diags(np.power(np.array(adj.sum(1)), -1).flatten())
        a_norm = d.dot(adj).tocsr()
    return a_norm

def preprocess_adj(adj, symmetric=True):
    adj = normalize_adj(adj, symmetric)
    return adj
```
