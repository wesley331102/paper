```python
def calculate_averages(list_2d):
  column_totals = dict()
  for row_idx, row in enumerate(list_2d):
    for cell_idx, cell in enumerate(row):
      if type(cell) in [int, float, complex]:
        if cell_idx in column_totals:
          column_totals[cell_idx].append(cell)
        else:
          column_totals[cell_idx] = [cell]
  per_col_avg = [sum(column_totals[col_idx]) / len(column_totals[col_idx]) for col_idx in column_totals]
  return per_col_avg
```

```python
def getWinRate(win, all):
  if all == 0:
    return 0.5
  else:
    return float(win) / all
```

```python
cool_dict = dict()
for i in range(len(team_list_y)):
  for key in team_list_y[i]:
    team1, team2 = key
    if (team1, team2) not in cool_dict and (team2, team1) not in cool_dict:
      cool_dict[key] = (0, 0)

for i in range(len(team_list_y)):
  for key in team_list_y[i]:
    team1, team2 = key
    flag = True
    if (team1, team2) in cool_dict:
      winRate = getWinRate(cool_dict[(team1, team2)][0], cool_dict[(team1, team2)][1])
      winScore = team_list_y[i][key]
      team_list_y[i][key] = (winScore, winRate)
    elif (team2, team1) in cool_dict:
      flag = False
      winRate = getWinRate(cool_dict[(team2, team1)][0], cool_dict[(team2, team1)][1])
      winScore = team_list_y[i][key]
      team_list_y[i][key] = (winScore, 1 - winRate)

    win, all = cool_dict[(team1, team2)] if flag else cool_dict[(team2, team1)]
    if team_list_y[i][key][0] > 0:
      if flag:
        cool_dict[(team1, team2)] = (win + 1, all + 1)
      else:
        cool_dict[(team1, team2)] = (win, all + 1)
    else:
      if flag:
        cool_dict[(team1, team2)] = (win, all + 1)
      else:
        cool_dict[(team1, team2)] = (win + 1, all + 1)
```

```python
# 21-22
start_date = '03/21/2022'
end_date = '04/15/2022'
date_list = [d.strftime('%m/%d/%Y') for d in pd.date_range(start=start_date, end=end_date)]
player_team_dict = pickle.load(open('average_dict.p', 'rb' ))
```

```python
p_list = list()
r_list = list()
a_list = list()

for err in error_dict.keys():
  for day in schedule_dict.keys():
    for match_ in schedule_dict[day]:
      z = list(match_.keys())
      if z[0] == team_index_dict[err[0]] and z[1] == team_index_dict[err[1]]:
        if (error_dict[err][0] > 0 and match_[z[0]] == 1) or (error_dict[err][0] < 0 and match_[z[0]] == 0):
          if day in date_list:
            cool_list = list()
            for all_day in player_team_dict.keys():
              if (err[0], err[1]) in player_team_dict[all_day]:
                cool_list.append(player_team_dict[all_day][(err[0], err[1])][0])
              elif (err[1], err[0]) in player_team_dict[all_day]:
                cool_list.append(-player_team_dict[all_day][(err[1], err[0])][0])
            ave = sum(cool_list) / len(cool_list)
            p_list.append(error_dict[err][1])
            r_list.append(error_dict[err][0])
            a_list.append(ave)
            print("{}_{}_vs_{}_predict_{}_result_{}_ave_{}".format(day, team_name_dict[err[0]], team_name_dict[err[1]], error_dict[err][1], error_dict[err][0], ave))
```
