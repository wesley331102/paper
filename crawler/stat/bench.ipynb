```python
s_list = pickle.load(open( "schedule4.p", "rb" ))
```

```python
def oppo_player(date, ssn, schedule_list): 

    headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36', 'Referer': 'https://www.nba.com/'}

    if not os.path.exists('./pos'):
        os.mkdir('./pos')
    
    date_name = date.replace('/', '-')
    if not os.path.exists(os.path.join('pos', date_name)):
        os.mkdir(os.path.join('pos', date_name))
        
    if not os.path.exists('./pos_all'):
        os.mkdir('./pos_all')
    
    date_name = date.replace('/', '-')
    if not os.path.exists(os.path.join('pos_all', date_name)):
        os.mkdir(os.path.join('pos_all', date_name))
        
    match_list = schedule_list[date]
    for match in match_list:
        time.sleep(0.5)
        team1 = list(match.keys())[0]
        team2 = list(match.keys())[1]
        url = 'https://stats.nba.com/stats/leaguedashptstats?College=&Conference=&Country=&DateFrom=' + str(date) + '&DateTo=' + str(date) + '&Division=&DraftPick=&DraftYear=&GameScope=&Height=&LastNGames=0&LeagueID=&Location=&Month=0&OpponentTeamID=0&Outcome=&PORound=&PerMode=Totals&PlayerExperience=&PlayerOrTeam=Player&PlayerPosition=&PtMeasureType=SpeedDistance&Season=' + str(ssn) + '&SeasonSegment=&SeasonType=Regular+Season&StarterBench=Starters&TeamID=' + str(team1) + '&VsConference=&VsDivision=&Weight='
        json = requests.get(url, headers=headers).json()
        data = json['resultSets'][0]['rowSet']
        columns = json['resultSets'][0]['headers']
        df = pd.DataFrame.from_records(data, columns=columns)
        df.to_csv(os.path.join('pos', date_name, str(team1) + '.csv'))
        time.sleep(0.5)
        url = 'https://stats.nba.com/stats/leaguedashptstats?College=&Conference=&Country=&DateFrom=' + str(date) + '&DateTo=' + str(date) + '&Division=&DraftPick=&DraftYear=&GameScope=&Height=&LastNGames=0&LeagueID=&Location=&Month=0&OpponentTeamID=0&Outcome=&PORound=&PerMode=Totals&PlayerExperience=&PlayerOrTeam=Player&PlayerPosition=&PtMeasureType=SpeedDistance&Season=' + str(ssn) + '&SeasonSegment=&SeasonType=Regular+Season&StarterBench=Starters&TeamID=' + str(team2) + '&VsConference=&VsDivision=&Weight='
        json = requests.get(url, headers=headers).json()
        data = json['resultSets'][0]['rowSet']
        columns = json['resultSets'][0]['headers']
        df = pd.DataFrame.from_records(data, columns=columns)
        df.to_csv(os.path.join('pos', date_name, str(team2) + '.csv'))
        time.sleep(0.5)
        url = 'https://stats.nba.com/stats/leaguedashptstats?College=&Conference=&Country=&DateFrom=' + str(date) + '&DateTo=' + str(date) + '&Division=&DraftPick=&DraftYear=&GameScope=&Height=&LastNGames=0&LeagueID=&Location=&Month=0&OpponentTeamID=0&Outcome=&PORound=&PerMode=Totals&PlayerExperience=&PlayerOrTeam=Player&PlayerPosition=&PtMeasureType=SpeedDistance&Season=' + str(ssn) + '&SeasonSegment=&SeasonType=Regular+Season&TeamID=' + str(team1) + '&VsConference=&VsDivision=&Weight='
        json = requests.get(url, headers=headers).json()
        data = json['resultSets'][0]['rowSet']
        columns = json['resultSets'][0]['headers']
        df = pd.DataFrame.from_records(data, columns=columns)
        df.to_csv(os.path.join('pos_all', date_name, str(team1) + '.csv'))
        time.sleep(0.5)
        url = 'https://stats.nba.com/stats/leaguedashptstats?College=&Conference=&Country=&DateFrom=' + str(date) + '&DateTo=' + str(date) + '&Division=&DraftPick=&DraftYear=&GameScope=&Height=&LastNGames=0&LeagueID=&Location=&Month=0&OpponentTeamID=0&Outcome=&PORound=&PerMode=Totals&PlayerExperience=&PlayerOrTeam=Player&PlayerPosition=&PtMeasureType=SpeedDistance&Season=' + str(ssn) + '&SeasonSegment=&SeasonType=Regular+Season&TeamID=' + str(team2) + '&VsConference=&VsDivision=&Weight='
        json = requests.get(url, headers=headers).json()
        data = json['resultSets'][0]['rowSet']
        columns = json['resultSets'][0]['headers']
        df = pd.DataFrame.from_records(data, columns=columns)
        df.to_csv(os.path.join('pos_all', date_name, str(team2) + '.csv'))
```
